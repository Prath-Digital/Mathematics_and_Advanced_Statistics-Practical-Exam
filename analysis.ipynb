## 🧱 Step 0: Dataset Creation

```python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 4000

df = pd.DataFrame({
    "Employee_ID": range(1, n+1),
    "Department": np.random.choice(["HR","IT","Sales","Finance"], n),
    "Age": np.random.randint(22,60,n),
    "Salary": np.random.randint(20000,100000,n),
    "Projects_Completed": np.random.randint(1,20,n),
    "Working_Hours": np.random.randint(6,12,n),
    "Performance_Score": np.random.randint(40,100,n)
})

df["Promotion_Status"] = np.where(df["Performance_Score"]>75,"Yes","No")
df.head()
```

## 📈 Step 1: Central Tendency

```python
print("Mean:", df["Salary"].mean())
print("Median:", df["Salary"].median())
print("Mode:", df["Salary"].mode()[0])
```

## 📊 Step 2: Variance & Standard Deviation

```python
print("Variance:", df["Projects_Completed"].var())
print("Std Dev:", df["Projects_Completed"].std())
```

## 🎲 Step 3: Probability of Promotion

```python
prob = (df["Promotion_Status"] == "Yes").mean()
print("Promotion Probability:", prob)
```

## 🔢 Step 4: Contingency Table

```python
table = pd.crosstab(df["Department"], df["Promotion_Status"])
print(table)
```

## 🎯 Step 5: Conditional Probability

```python
high_perf = df[df["Performance_Score"] > 80]
cond_prob = (high_perf["Promotion_Status"] == "Yes").mean()
print("Conditional Probability:", cond_prob)
```

## 📉 Step 6: Histogram + Gaussian Curve

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Performance_Score"], kde=True)
plt.title("Performance Score Distribution")
plt.show()
```

## 📏 Step 7: Skewness & Kurtosis

```python
print("Skewness:", df["Salary"].skew())
print("Kurtosis:", df["Salary"].kurt())
```

## 📐 Step 8: Q-Q Plot

```python
from scipy import stats

stats.probplot(df["Projects_Completed"], dist="norm", plot=plt)
plt.show()
```

## 🔢 Step 9: Linear Algebra

```python
vectors = df[["Projects_Completed", "Working_Hours"]].head(5).values

v1 = vectors[0]
v2 = vectors[1]

dot_product = np.dot(v1, v2)
print("Dot Product:", dot_product)

norm1 = np.linalg.norm(v1)
norm2 = np.linalg.norm(v2)

print("Norm1:", norm1)
print("Norm2:", norm2)

angle = np.arccos(dot_product / (norm1 * norm2))
print("Angle (radians):", angle)
```
